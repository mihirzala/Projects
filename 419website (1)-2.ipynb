```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# Getting links of the mails
url = 'https://www.419scam.org/emails/'
resp = requests.get(url)
soup = BeautifulSoup(resp.content, features = 'html')
print(soup)
```

```python
extract_link = []
all_links = soup.find_all('a')
for link in all_links:
    if len(link.text.strip()) == 2:
        extract_link.append(url+str(link.get('href')))

print(extract_link)
```
